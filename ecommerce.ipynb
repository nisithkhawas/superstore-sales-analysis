```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
import kaleido
```

```python
data = pd.read_csv("data/Sample - Superstore.csv",encoding='latin-1')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

Converting date column

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Day of Week'] = data['Order Date'].dt.day
data['Order Year'] = data['Order Date'].dt.year
```

# Monthly sales analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
fig = px.line(
    sales_by_month,
    x='Order Month',
    y='Sales',
    title='Monthly Sales Analysis'
)

fig.show()

fig.write_image("images/monthly_sales.png")
```

```python
data.head()
```

# sales by category

```python
sales_by_category = data.groupby(['Category'])['Sales'].sum().reset_index()
sales_by_category
```

```python
fig = px.pie(
    sales_by_category,
    values='Sales',
    names='Category',
    hole=0.5,
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label',
    insidetextorientation='radial'
)

fig.update_layout(
    title_text='Sales Analysis by Category',
    title_font=dict(size=24),
    showlegend=False   # side rectangle hat gaya
)

fig.show()

fig.write_image("images/sales_by_category.png")
```

# sales analysis by sub category

```python
data.head()
```

```python
sales_by_subcategory = data.groupby(data['Sub-Category'])['Sales'].sum().reset_index()
sales_by_subcategory
```

```python
fig = px.bar(
    sales_by_subcategory,
    x='Sub-Category',
    y='Sales',
    title='Sales Analysis by Sub Category'
)

fig.show()
fig.write_image("images/sales_by_subcategory.png")
```

# sales and profit - customer segment

```python
data.head()
```

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment
```

```python
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment',yaxis_title='Amount')

fig.show()

fig.write_image("images/sales_and_profit_by_customerSegment.png")

```

# sales to profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Sales','Sales_to_Profit_Ratio']])
```
